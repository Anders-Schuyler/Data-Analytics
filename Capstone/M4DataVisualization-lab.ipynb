# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of ConvertedComp.

```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the ConvertedComp column
query = "SELECT ConvertedComp FROM master"
df = pd.read_sql_query(query, conn)

# Drop missing values from the 'ConvertedComp' column
converted_comp = df['ConvertedComp'].dropna()

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(converted_comp, bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribution of Converted Compensation (USD)')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Close the connection
conn.close()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the Age column
query = "SELECT Age FROM master"
df = pd.read_sql_query(query, conn)

# Drop missing values from the 'Age' column
age_data = df['Age'].dropna()

# Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(age_data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()

# Close the connection
conn.close()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the Age and WorkWeekHrs columns
query = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(query, conn)

# Drop rows with missing values in either Age or WorkWeekHrs
scatter_data = df.dropna(subset=['Age', 'WorkWeekHrs'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(scatter_data['Age'], scatter_data['WorkWeekHrs'], alpha=0.6, edgecolor='black')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()

# Close the connection
conn.close()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the relevant columns
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(query, conn)

# Drop rows with missing values in either WorkWeekHrs, CodeRevHrs, or Age
bubble_data = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Plot the bubble plot
plt.figure(figsize=(10, 6))
bubble_size = bubble_data['Age']  # Bubble size is based on Age
plt.scatter(bubble_data['WorkWeekHrs'], bubble_data['CodeRevHrs'], 
            s=bubble_size, alpha=0.5, edgecolor='w')

plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble size = Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)

# Show the plot
plt.show()

# Close the connection
conn.close()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the relevant column for databases
query = "SELECT DatabaseDesireNextYear FROM DatabaseDesireNextYear"
df = pd.read_sql_query(query, conn)

# Split the entries by commas (since a respondent can select multiple databases)
df_split = df['DatabaseDesireNextYear'].str.split(';', expand=True)

# Unpivot the dataframe to have a single column of all databases
df_unpivoted = df_split.melt(value_name='Database').dropna()

# Count the number of occurrences for each database
database_counts = df_unpivoted['Database'].value_counts().head(5)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(database_counts, labels=database_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

# Close the connection
conn.close()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve Age, WorkWeekHrs, and CodeRevHrs
query = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age BETWEEN 30 AND 35"
df = pd.read_sql_query(query, conn)

# Drop rows with missing values in WorkWeekHrs and CodeRevHrs
df_clean = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs'])

# Calculate the median WorkWeekHrs and CodeRevHrs for age group 30 to 35
median_workweekhrs = df_clean['WorkWeekHrs'].median()
median_coderevhrs = df_clean['CodeRevHrs'].median()

# Data for the stacked bar chart
data = {'WorkWeekHrs': median_workweekhrs, 'CodeRevHrs': median_coderevhrs}
categories = list(data.keys())
values = list(data.values())

# Plotting the stacked bar chart
plt.figure(figsize=(8, 6))
plt.barh(categories, values, color=['#1f77b4', '#ff7f0e'])
plt.title('Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.xlabel('Hours')
plt.ylabel('Activity')
plt.show()

# Close the connection
conn.close()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve Age and ConvertedComp for respondents aged 45 to 60
query = "SELECT Age, ConvertedComp FROM master WHERE Age BETWEEN 45 AND 60"
df = pd.read_sql_query(query, conn)

# Drop rows with missing values in ConvertedComp
df_clean = df.dropna(subset=['ConvertedComp'])

# Group by Age and calculate the median ConvertedComp for each age
median_convertedcomp = df_clean.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(10, 6))
plt.plot(median_convertedcomp['Age'], median_convertedcomp['ConvertedComp'], marker='o', color='b')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.show()

# Close the connection
conn.close()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
conn = sqlite3.connect("m4_survey_data.sqlite")
# Query to retrieve the MainBranch column
query = "SELECT MainBranch FROM master"
df = pd.read_sql_query(query, conn)

# Count the occurrences of each unique value in the MainBranch column
main_branch_counts = df['MainBranch'].value_counts()

# Plot a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(True)
plt.show()

# Close the connection
conn.close()
```
